```python
from transformers import pipeline

# Load sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# Input text
text = "I really enjoyed this product. It works perfectly!"

# Predict sentiment
result = classifier(text)

print(result)
```

```python
classifier = pipeline("sentiment-analysis")

reviews = [
    "This phone is excellent.",
    "The battery life is terrible.",
    "The product is okay."
]

results = classifier(reviews)

for review, result in zip(reviews, results):
    print(f"Review: {review}")
    print(f"Sentiment: {result['label']}")
    print(f"Confidence: {result['score']:.4f}")
    print()
```

```python
import pandas as pd
from transformers import pipeline

# Load data
df = pd.read_csv("review.csv")

# Load model
classifier = pipeline("sentiment-analysis")

# Predict
df["sentiment"] = df["review"].apply(
    lambda x: classifier(x)[0]["label"]
)

print(df)
```

```python
reviews = df["review"].tolist()

results = classifier(
    reviews,
    batch_size=32,
    truncation=True
)

df["sentiment"] = [r["label"] for r in results]
```

```python
df = pd.read_csv("review.csv")

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

results = classifier(
    df["review"].tolist(),
    batch_size=16,
    truncation=True
)

df["sentiment"] = [r["label"] for r in results]
df["confidence"] = [r["score"] for r in results]

df.to_csv("output.csv", index=False)
```

```python
from gtts import gTTS
import whisper
from transformers import pipeline

input_review = input("Enter the user review:")
# text = "I had an issue with how the subscription was presented after significant amount was deducted from my account.I am happy to report that I was promptly sorted. This was top notch customer service for audio team! I recommend them"
raw_audio = gTTS(text=input_review,lang="en")
raw_audio.save("raw.mp3")
model = whisper.load_model("base")
text_result = model.transcribe("raw.mp3")
analyzer = pipeline("sentiment-analysis")
analysis_result = analyzer(text_result['text'])
if analysis_result[0]['label']=='POSITIVE':
    audio_result = gTTS(text=f"User gives {analysis_result[0]['label']} with a score of {int((analysis_result[0]['score'])*100)} percent",lang="en").save("result.mp3")
    from IPython.display import display,Audio
    display(Audio("result.mp3"))
else:
    print("User gives Negative review")
```

```python
import pandas as pd
from transformers import pipeline

# Load the amazon.csv file into a DataFrame
df_amazon = pd.read_csv("amazon.csv")

# Initialize the sentiment analysis pipeline (if not already initialized)
# Re-initializing here to ensure it's available, though it might be redundant if the previous cell was run.
classifier = pipeline("sentiment-analysis")

# Clean the 'reviewText' column: explicitly convert all entries to string and handle NaNs
df_amazon["reviewText"] = df_amazon["reviewText"].apply(lambda x: str(x) if pd.notna(x) else '')

# Perform sentiment analysis on the 'reviewText' column
reviews_list = df_amazon["reviewText"].tolist()
results_amazon = classifier(
    reviews_list,
    batch_size=32, # Use batch processing for efficiency
    truncation=True # Truncate long texts if necessary
)

# Extract sentiment labels and confidence scores
df_amazon["sentiment"] = [r["label"] for r in results_amazon]
df_amazon["confidence"] = [r["score"] for r in results_amazon]

# Display the DataFrame with sentiment analysis results
display(df_amazon.head())
```
